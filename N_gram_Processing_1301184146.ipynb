```python
import pandas as pd
import re
import string
import nltk
```

```python
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()

data = pd.read_csv("/content/stock_data - stock_data.csv")
# data.columns= ['Text','Sentiment']
data.head()
# data.shape
```

## Text Cleaning

```python
def clean_text (text):
    text = "".join([c for c in text if c not in string.punctuation]) #menghapus tanda baca pada kalimat
#     print(text)
    token = re.split('\W+', text) #split kata
    text = " ".join([ps.stem(word) for word in token if word not in stopwords])
#     print(token)
    return text
```

```python
string.punctuation
```

```python
data['cleaned_sentence'] = data['Text'].apply(lambda x: clean_text(x)) #mengembalikan data["message"] dari fungsi namun data["message"] tidak berubah
data.head()
```

## Count Vectorizer
#### 1. import library sklearn untuk mengambil Count Vectorizer
#### 2. buat corpus buatan
#### 3. hitung jumlah vocabularu dengan count vectorizer --> CountVectorizer().fit(corpus)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range = (2,2))

corpus = ["One of the example",
         "This is an example",
         "Just example of sentence"]

X = cv.fit(corpus)
print(X.vocabulary_)

print(cv.get_feature_names())
```

```python
X = cv.transform(corpus)

print(X.shape)
# print(X)

print(X.toarray())
arrx = X.toarray()

df =  pd.DataFrame(arrx, columns = cv.get_feature_names())
print(df)
```
